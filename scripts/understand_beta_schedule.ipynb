```python
import numpy as np
import matplotlib.pyplot as plt


fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(15, 10))

for n_steps in (100, 500, 1_000, 10_000):
    beta = np.linspace(1e-4, 0.02, n_steps)
    alpha = 1 - beta
    alpha_bar = np.cumprod(alpha)
    noise_multiplier = np.sqrt(1 - alpha_bar) 
    prev_multiplier = np.sqrt(alpha_bar)
    axs[0][0].plot(np.arange(1, n_steps + 1) / n_steps, noise_multiplier, label=n_steps)
    axs[1][0].plot(np.arange(1, n_steps + 1) / n_steps, prev_multiplier, label=n_steps)
    axs[0][1].plot(np.arange(1, n_steps + 1) / n_steps, beta / noise_multiplier, label=n_steps)
    axs[1][1].plot(np.arange(1, n_steps + 1) / n_steps, beta, label=n_steps)
    axs[2][0].plot(np.arange(1, n_steps + 1) / n_steps, 1 / np.sqrt(alpha), label=n_steps)

axs[0][0].legend()
axs[0][0].set_title("Noise multiplier - noising")
axs[1][0].set_title("X_0 multiplier - noising")
axs[0][1].set_title("Noise multiplier - denoising")
axs[1][1].set_title("Beta - variance of sampling dist - denoising")
axs[2][0].set_title("Scaler of prev - eps * noise for denoising mean")

plt.show()
```
